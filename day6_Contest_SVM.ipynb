```python
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

### 실습 과제 1. 데이터 불러온 후 NaN 값이 있다면 해당 행 제거하고, train_data : test_data = 7:3 split
#### 종속변수는 PKT_Y, 나머지 변수들이 독립변수

```python
df = pd.read_csv('./data/contest.csv')
df.head()
```

```python
print(f"df.shape: {df.shape}")
print()

df.info()
```

```python
df.describe(include='all')
```

```python
num_null = df.isnull().sum()
num_null[num_null != 0]
```

```python
df = df.dropna(subset=['PKT_Y'], how='any')
```

```python
df.shape
```

```python
X = df[df.columns[6:-1]]
Y = df['PKT_Y']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state=10)

print(f"train dataset size: {x_train.shape}")
print(f"test dataset size: {x_test.shape}")
```

### 실습 과제 2. 선형 커널, 3차항 커널, RBF 커널, 시그모이드 커널로 각각 SVM 모델 구현 후 accuracy, F1 score, confusion matrix, precision, recall (C=1, gamma = 0.1)

```python
svm_linear = svm.SVC(kernel='poly', degree=1, C=1, gamma=0.1)
svm_poly = svm.SVC(kernel='poly', degree=3, C=1, gamma=0.1)
svm_rbf = svm.SVC(kernel='rbf', C=1, gamma=0.1)
svm_sig = svm.SVC(kernel='sigmoid', C=1, gamma=0.1)

svm_model_names = ['linear', 'poly', 'rbf', 'sig']
svm_models = [svm_linear, svm_poly, svm_rbf, svm_sig]

for name, model in zip(svm_model_names, svm_models):
    print(f"model name is svm_{name}")
    print('-'*50)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(f"accuracy: {metrics.accuracy_score(y_test, y_pred)}")
    print(f"f1_score: {metrics.f1_score(y_test, y_pred)}")
    print(f"precision_score: {metrics.precision_score(y_test, y_pred)}")
    print(f"recall_score: {metrics.recall_score(y_test, y_pred)}")
    print(f"confusion_mtrx: \n{metrics.confusion_matrix(y_test, y_pred)}")
    print()
```

### 실습 과제 3. 하이퍼 파라미터 C 값(0.01, 0.1, 1, 10)에 따른 RBF 커널을 SVM 모델의 accuracy와 F1 score 변화 관찰

```python
C_list = [0.01, 0.1, 1, 10]

for c_value in C_list:
    svm_rbf = svm.SVC(kernel='rbf', C=c_value, gamma=0.1)
    svm_rbf.fit(x_train, y_train)
    y_pred = svm_rbf.predict(x_test)
    
    print(f"c value: {c_value}")
    print(f"accuracy: {metrics.accuracy_score(y_test, y_pred)}")
    print(f"f1_score: {metrics.f1_score(y_test, y_pred)}")
    print('-'*30)
```

### 실습 과제 4. 하이퍼 파라미터 gamma 값(0.005,0.01,0.1, 1)에 따른 RBF 커널을 SVM 모델의 accuracy와 F1 score 변화 관찰

```python
gamma_list = [0.005, 0.01, 0.1, 1]

for gamma_value in gamma_list:
    svm_rbf = svm.SVC(kernel='rbf', C=1, gamma=gamma_value)
    svm_rbf.fit(x_train, y_train)
    y_pred = svm_rbf.predict(x_test)
    
    print(f"gamma value: {gamma_value}")
    print(f"accuracy: {metrics.accuracy_score(y_test, y_pred)}")
    print(f"f1_score: {metrics.f1_score(y_test, y_pred)}")
    print('-'*30)
```

### 실습 과제 5. 하이퍼 파라미터 degree(1,2,3,4,5)에 따른 다항식 커널 SVM 모델 accuracy와 F1 score 변화 관찰

```python
degree_list = [1,2,3,4,5]

for value in degree_list:
    svm_poly = svm.SVC(kernel='poly', degree=value, C=1, gamma=0.1)
    svm_poly.fit(x_train, y_train)
    y_pred = svm_poly.predict(x_test)
    
    print(f"degree: {value}")
    print(f"accuracy: {metrics.accuracy_score(y_test, y_pred)}")
    print(f"f1_score: {metrics.f1_score(y_test, y_pred)}")
    print('-'*30)
```
