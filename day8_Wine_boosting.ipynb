```python
# 16. XGBoosting 모델의 파라미터 탐색을 위한 함수를 정의하라.
def parameter_search(params, random=False):
```

```python
#16. your code here
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

def parameter_search(params, random=False):
    xgb = XGBClassifier(booster='gbtree', objective='binary:logistic',
                        random_state=11, verbosity=0, use_label_encoder=False)
    kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=11)
    
    if random:
        grid = RandomizedSearchCV(xgb, params, cv=kfold, n_iter=5, n_jobs=-1, random_state=11)
    else:
        grid = GridSearchCV(xgb, params, cv=kfold, n_jobs=-1)
        
    grid.fit(X_train, y_train)
    print(f"최적의 매개변수: {grid.best_params_}")
    print(f"최적의 점수: {grid.best_score_:.5}")
```

```python
# 17. GridSearchCV 방법을 통해서 다음과 같은 파라미터들에 대해서 최적의 조합을 찾아라.
# 탐색을 하기위한 시간도 함께 출력하라.
# n_estimators:[10, 20, 40] /  'max_depth': [10, 20,40] / 'learning_rate': [0.1,0.2,0.5]
```

```python
#17. your code here
from time import time

params = {'n_estimators':[10, 20, 40], 'max_depth':[10, 20,40], 'learning_rate':[0.1,0.2,0.5]}

start = time()
parameter_search(params)
end = time()

print(f"GridSearchCV: {end - start:.5f} sec")
```

```python
# 18. RandomSearchCV 방법을 통해서 다음과 같은 파라미터들에 대해서 최적의 조합을 찾아라.
# 탐색을 하기위한 시간도 함께 출력하라.
# n_estimators:[10, 20, 40] /  'max_depth': [10, 20,40] / 'learning_rate': [0.1,0.2,0.5]
```

```python
#18. your code here
params = {'n_estimators':[10, 20, 40], 'max_depth':[10, 20,40], 'learning_rate':[0.1,0.2,0.5]}

start = time()
parameter_search(params, random=True)
end = time()

print(f"RandomizedSearchCV: {end - start:.5f} sec")
```
